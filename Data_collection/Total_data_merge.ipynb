```python
# from my_func import *
import pandas as pd
import numpy as np
import os
```

```python
parquet_list = os.listdir('/Users/packchansaem/Desktop/project/maket_kurly/parquet_data/')
parquet_list.sort()
parquet_list
```

```python
parquet_list.pop(0)
```

```python
a_list = ['무', '배추', '대파', '건고추', '마늘', '양파','사과']
df_total = pd.DataFrame()
for i in parquet_list:
    df = pd.read_parquet(f'parquet_data/{i}')
    selcet_df=df[df['PUM_NM'].apply(lambda x: x in a_list)]
    df_total = pd.concat([df_total,selcet_df],axis=0)
```

```python
df_total
```

```python
df_total.info()
```

```python
#메모리 최소화를 위해 카테고리 타입으로 변경
df_total=df_total.astype({'KIND_NM':'category','CMP_NM':'category','LV_NM':'category','SAN_NM':'category','PUM_NM':'category'})
df_total=df_total.astype({'WHSAL_NM':'category','DAN_NM':'category','POJ_NM':'category','SIZE_NM':'category'})

#결측치 채우기
df_total['LV_NM']=df_total['LV_NM'].fillna('.')
df_total['POJ_NM']=df_total['POJ_NM'].fillna('.')
df_total['SIZE_NM']=df_total['SIZE_NM'].fillna('.')
df_total['SAN_NM']=df_total['SAN_NM'].astype('object')
df_total['SAN_NM']=df_total['SAN_NM'].fillna('.')
df_total['SAN_NM']=df_total['SAN_NM'].astype('category')
df_total['DAN_NM']=df_total['DAN_NM'].fillna('.')

#데이터타입을 datetime으로 변경
df_total["SALEDATE"] = pd.to_datetime( df_total["SALEDATE"], format='%Y%m%d' )

#거래가 취소되거나 거래량이 0인 경우 드랍
df_total=df_total.drop(df_total[df_total['TOT_QTY']<=0].index)
df_total=df_total.drop(df_total[df_total['TOT_AMT']<=0].index)

#index 정렬
df_total = df_total.sort_values('SALEDATE').reset_index(drop=True)
```
