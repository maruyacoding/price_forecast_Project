## 경기도
- 기상 데이터의 경기도 관측소 ['동두천', '수원', '양평', '이천', '파주']

```python
서울_lst = ['고양', '구리', '남양주', '김포', '하남', '의정부', '성남', '광명', '과천']
수원_lst = ['안성', '안산', '용인', '화성', '평택', '시흥', '안양', '오산', '의왕', '군포']
동두천_lst = ['포천', '연천', '양주']
양평_lst = ['광주', '가평']
이천_lst = ['여주']
```

```python
print('---변경 전 확인---')
for i in ['서울', '경기'] :
    display(df1[df1.SAN_NM.str.contains(i)]['SAN_NM'].value_counts())
    display(df1[df1.SAN_NM.str.contains(i)]['SAN_NM'].nunique())


print('---산지명 변경---')
for i in 서울_lst :
    df1.loc[df1.SAN_NM.str.contains(i), 'SAN_NM'] = '서울' #
for i in 수원_lst :
    df1.loc[df1.SAN_NM.str.contains(i), 'SAN_NM'] = '경기도 수원'
for i in 동두천_lst :
    df1.loc[df1.SAN_NM.str.contains(i), 'SAN_NM'] = '경기도 동두천'
for i in 양평_lst :
    df1.loc[df1.SAN_NM.str.contains(i), 'SAN_NM'] = '경기도 양평'
for i in 이천_lst :
    df1.loc[df1.SAN_NM.str.contains(i), 'SAN_NM'] = '경기도 이천' 

df1.loc[df1.SAN_NM.str.contains('부천'), 'SAN_NM'] = '인천' #
 


print('---변경 후 확인---')
for i in ['서울', '경기'] :
    display(df1[df1.SAN_NM.str.contains(i)]['SAN_NM'].value_counts())
    display(df1[df1.SAN_NM.str.contains(i)]['SAN_NM'].nunique())
```
