### 광역시 처리

```python
# 광주광역시, 대구광역시, 대전광역시, 부산광역시, 세종시, 울산광역시, 인천광역시

# 경기도 광주는 관측소 없음, 따라서 광주는 광주광역시의 관측소를 말하는 것!

df1.loc[df1.지점명.str.contains('광주'), '지점명'] = '광주광역시'
df1.loc[df1.지점명.str.contains('대구'), '지점명'] = '대구광역시'
df1.loc[df1.지점명.str.contains('대전'), '지점명'] = '대전광역시'
df1.loc[df1.지점명.str.contains('부산'), '지점명'] = '부산광역시'
df1.loc[df1.지점명.str.contains('세종'), '지점명'] = '세종시'
df1.loc[df1.지점명.str.contains('울산'), '지점명'] = '울산광역시'
df1.loc[df1.지점명.str.contains('인천'), '지점명'] = '인천'


df1.loc[df1.지점명.str.contains('강화'), '지점명'] = '인천 강화도'
```
